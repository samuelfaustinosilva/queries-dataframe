### Combinando as informações dos dataframes com linguagem SQL

```python
query = """

SELECT
p.id as product_id,
p.name as product_name,
p.price,
c.name as category_name 

FROM products p
JOIN categories c on c.id = p.category_id

"""
```

```python
resultado = sqldf(query)
```
